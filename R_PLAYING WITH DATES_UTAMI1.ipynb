```r
library(dplyr)
```

open a package of data manipulation in Dplyr. Dplyr refers to a grammar of data manipulation, as quoted from Hadley Wickham. Dplyr has three main function, one of which is to identify the most important data manipulation verbs and make them easy to use from R. Command that can be found in dplyr: mutate (), select (), filter (), summarise () and arrange (). 

```r
lpdassaults <- read.csv("Data/lpdassaults.csv", stringsAsFactors = FALSE)
```

import the data of lpd assaults. add stringAsFactors = False to avoid R sees date as numbers. 

```r
head (lpdassaults)
```

overal header

```r
lpdassaults %>% mutate(
    CleanDate = as.POSIXct(date, format="%Y-%m-%dT%H:%M"),
)
```

So, because it is weird the format of the date with a T in the middle of it>> 2010-04-05T00:00:00.000Z >> year-month-day T hours-minutes-seconds or YYYY-MM-DDTHH:MM:SS, we use mutate command. Mutate, as mentioned by Hadley Wickham, has function to adds new variables that are functions of existing variables. 

CleanDate is the new field and we use asPOSIXct because as.POSIXct uses date and time with associated time zone. It also convert a date-time string into POSIXct class. 

because we want to know what month that the assaults incidents toward lpd officers mostly occured, we need to parse the pattern of YYYY-MM-DDTHH:MM:SS by using characters as explained in https://stat.ethz.ch/R-manual/R-devel/library/base/html/strptime.html

    So, the pattern would be "%Y-%m-%dT%H:%M"

```r
lpdassaults %>% mutate(
    month = format(as.POSIXct(date, format="%Y-%m-%dT%H:%M"), "%m"),
)
```

mutate again and use the new field month to explain the pattern "%Y-%m-%dT%H:%M" (YYYY-MM-DDTHH:MM:SS). Take out "%m" since we want to know the what month and not what year. 

```r
lpdassaults %>% mutate(
    month = format(as.POSIXct(date, format="%Y-%m-%dT%H:%M"), "%m"),
) %>% group_by(month) %>% summarize(
    count = n()
)
```

count them on!

the most assault to the LPD officers were happened in April and November, which reached to 36 and 36 incidents, respectively. 

Step:

To figure out the month, use the character '%m' as it refers to month. then, do the group by and count how many assaults happened in each month.
