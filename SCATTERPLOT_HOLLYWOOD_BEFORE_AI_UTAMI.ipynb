```r
library(ggplot2)
```

IMPORT GGPLOT

```r
movies <- read.csv("Data/movies.csv")
```

import the data

```r
head(movies)
```

see overal look of the data

```r
ggplot(movies, aes(x=Actor.1.Age, y=Actor.2.Age)) + geom_point()
```

call out the scatter plot!

```r
ggplot(movies, aes(x=Actor.1.Age, y=Actor.2.Age)) + 
geom_point() + 
geom_text(data=movies, aes(label=Movie.Name))
```

label the scatter plots!

```r
ggplot(movies, aes(x=Actor.1.Age, y=Actor.2.Age)) + 
geom_point() + 
geom_text(data=subset(movies, Age.Difference>=30), aes(label=Movie.Name))
```

limit the data so it wont be overlapped

```r
ggplot(movies, aes(x=Actor.1.Age, y=Actor.2.Age)) + 
geom_point() + 
geom_text(data=subset(movies, Age.Difference>=30), aes(label=Movie.Name), size=2, nudge_y=1, check_overlap = TRUE) +
labs(x="Actor 1 Age", y="Actor 2 Age", title="The age gap in Hollywood movie relationships")
```

name the scatter plot, move the label, resize, etc

```r
ggplot(movies, aes(x=Actor.1.Age, y=Actor.2.Age, color = "red")) + 
geom_point() + 
geom_smooth(span = 0.5) + 
geom_text(data=subset(movies, Age.Difference>35), aes(label=Movie.Name), size=2, nudge_y=1, check_overlap = TRUE) +
labs(x="Actor 1 Age", y="Actor 2 Age", title="The age gap in Hollywood movie relationships")
```

give some colors!

```r
ggplot(movies, aes(x=Actor.1.Age, y=Actor.2.Age, color = "red")) + 
geom_point (size = 0.5, shape = 8) +
geom_smooth(span = 0.5, color ="maroon")+ 
geom_text(data=subset(movies, Age.Difference>35), aes(label=Movie.Name), size=2, nudge_y=1.5, check_overlap = TRUE) +
labs(x="Actor 1 Age", y="Actor 2 Age", title="The age gap in Hollywood movie relationships", subtitle="The biggest difference found in Harold and Maude with 52 years age gap", caption="Source: Hollywoodagegap.com | By Utami")+
theme(plot.title = element_text(size = 16, face = "bold"), axis.title = element_text(size = 10), axis.ticks = element_blank())
```

change the colors of the line, put subtitle, source and byline
