```r
library(dplyr)
```

```r
NUsalaries <- read.csv("Data/nusalaries1718-3.csv")
```

```r
head(NUsalaries)
```

```r
NUsalaries %>% filter(Campus == "UNL")
```

```r
NUsalaries %>%
  group_by(Position) %>%
  summarise(
    count = n(),
  )
```

```r
NUsalaries %>%
  group_by(Position) %>%
  summarise(
    count = n(),
  ) %>% arrange(desc(count))
```

```r
NUsalaries %>% filter(Campus == "UNL")
```

```r
NUsalaries %>% 
    group_by(Position) %>%
    summarise(count = n (),
             Salariesmed = median(Budgeted.Annual.Salary),
             ) %>%
    arrange(desc(salariesmed))
```

```r
NUsalaries %>%
  summarise(
    count = n(),
    Salariesmean = mean(Budgeted.Annual.Salary),
      ) %>%
    arrange(desc(Salariesmed))
```
