```r
NUsalaries %>%
  group_by(Position) %>%
  summarise(
    count = n(),
      Salariesmed = median(Budgeted.Annual.Salary),
             ) %>%
    arrange(desc(Salariesmed))
```

The command is read: 1. take the data from NUsalaries dataset. 2. Do the group_by for (header) Position (because we want to calculate mean and median from job title)3. Count them up, using summarise because in dplyr we use summarise to count. 4. n? refers to a new field as in count equal to n() and 5. calculate the mean. choose the budgeted annual salary since two others headers state aided fund and from other funds have N/A which cannot be counted. 6. arrange in descending order to find the top median salaries. you find 5 top job title with highest median salaries: HEAD COACH-FOOTBALL, HEAD COACH-BASKETBALL, DIRECTOR OF ATHLETICS, ASSISTANT COACH FOOTBALL, HEAD COACH VOLLEYBALL.
