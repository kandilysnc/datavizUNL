need to group by incidents and count to figure out what is the most common incident reported to the unl police during 2013 - 2016

```r
unlcrime %>%
  group_by(Incident.Code) %>%
  summarise(
    count = n(),
  )
```

these are the incidents and their reports numbers to the police in random order

```r
unlcrime %>%
  group_by(Incident.Code) %>%
  summarise(
    count = n(),
  ) %>% arrange(desc(count))
```
