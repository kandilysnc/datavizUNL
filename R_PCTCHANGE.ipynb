```r
NEvoters %>% mutate(
  change = ((Total16 - Total10)/Total10)*100,
    repchange = ((Republican16 - Republican10)/Republican10)*100,
demchange = ((Democrat16 - Democrat10)/Democrat10)*100,
nonpartisanchange = ((Nonpartisan16 - Nonpartisan10)/Nonpartisan10)*100,
) %>% arrange(desc(change))
```

calculate percent change of the total registered voters in Nebraska from 2010 to 2016, the Republican Pct change, Dem Pct Change and Non Partisan Pct Change. Formulae: New-Old/Old *100. After that make it into big to small order by using %>% (des(change))

```r
NEvoters %>% mutate(
    repchange = ((Republican16 - Republican10)/Republican10)*100,
    ) %>% arrange(desc(repchange))
```

since we hardly see all the percent change in the box above. I will calculate pct change one by one to know the pct change for each party. As we can see above, the five top counties with the biggest percent change of Republican voters are: DIXON (20%), SARPY(15%), CEDAR(14%), SAUNDERS (12%) and GRANT (10%). 

```r
NEvoters %>% mutate(
    Nonpartisanchange = ((Nonpartisan16 - Nonpartisan10)/Nonpartisan10)*100,
    ) %>% arrange(desc(Nonpartisanchange))
```

As we can see above, the five top counties with the biggest percent change of Nonpartisan voters are: MCPHERSON (60%), ARTHUR(50%), THOMAS(44%), KEYA PAHA(42%), and BANNER(37%). 

```r
NEvoters %>% mutate(
    Demchange = ((Democrat16 - Democrat10)/Democrat10)*100,
    ) %>% arrange(desc(Demchange))
```

As we can see above, the five top counties with the biggest percent change of Democrat voters are: SARPY (9%), DOUGLAS(6%), THURSTON(3%), LANCASTER (1%),and DAWSON(1%).
